# Time resampling

```python
# https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78
TSLA_Data.resample(rule="A") # A- year end frequency
```

```python
TSLA_Data.resample(rule="A").min()
```

```python
TSLA_Data.resample(rule="A").max()
```

```python
TSLA_Data.resample(rule="A").max()["Open"].plot()
```

```python
TSLA_Data.resample(rule="BA").max()
```

```python
TSLA_Data.resample(rule="QS").max()
```

```python
TSLA_Data["Open"].resample(rule="A").mean().plot(kind="bar")
```

```python
TSLA_Data["Open"].resample(rule="M").mean().plot(kind="bar",figsize=(25,5))
```

```python
TSLA_Data["Open"].rolling(10)
```

```python
TSLA_Data["Open"].head(30).rolling(10).mean().plot()
```

```python
TSLA_Data["Open: 30 Days rolling"]=TSLA_Data["Open"].rolling(30).mean()
```

```python
TSLA_Data
```

```python
TSLA_Data[["Open","Open: 30 Days rolling"]].plot(figsize=(20,5))
```
